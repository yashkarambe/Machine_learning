# Importing the Libary

```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
```

* Data loading 

```python

x = np.array([[6.1 ], [6.15], [ 6.3] , [7.24] ,[7.5] ,[7.5], [ 7.9], [ 8.] , [8.9] ,[ 9.1 ] ,[ 9.5], [9.5] , [9.52]])
y = np.array([[0. ] ,[2.5] , [2.25] , [6.] , [3.3] , [3.75] , [4.5 ] , [3.3] , [4.] , [3.5]  , [6.5] , [10.5] , [ 14.5] ])
```

* Spliting of data

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state=42)
```

* Data visulization 

```python
plt.scatter(x_train,y_train , )
plt.xlabel("CGPA")
plt.ylabel("Salary")
plt.grid()
plt.plot()
```

* Model selection 

```python
linear_model = linear_model.LinearRegression()
poly_model = PolynomialFeatures(degree=3)
```

* Transform the linear data into the polynomial data

```python
x_train_trans = poly_model.fit_transform(x_train)
x_test_trans = poly_model.transform(x_test)
x_train_trans[0]
```

* Train , Test & calculate the score of Model

```python
linear_model.fit(x_train_trans , y_train)
x_pred = linear_model.predict(x_test_trans)
print(f'original value : {y_test[0]}')
print(f'predicted value : {x_pred[0]}')
r2_score( y_test , x_pred)
```

* Visulize the fit-Curve on the graph 

```python
plt.scatter(x_train,y_train)
plt.scatter(x_test , y_test)
plt.plot( x , linear_model.predict(poly_model.fit_transform(x)))
plt.xlabel("CGPA")
plt.ylabel("Salary")
plt.grid()
plt.plot()
```
