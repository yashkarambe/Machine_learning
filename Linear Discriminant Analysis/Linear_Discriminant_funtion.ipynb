# Importing library

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score , confusion_matrix
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt 
from mlxtend.plotting import plot_decision_regions
```

* Making Data Set 

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1)
```

* n_samples: Number of samples or data points to generate. In this case, it's set to 100,
             which means the dataset will have 100 data points.
* n_features: Number of features (or attributes) for each data point. Here, it's set to 2,
              which means each data point will have 2 features.
* n_informative: Number of informative features. Informative features are the ones that 
                 contribute to the class separation. In this case, it's set to 2, meaning 
                 both features will be informative.
* n_redundant: Number of redundant features. Redundant features are correlated with informative
               features and do not provide additional information. Here, it's set to 0, so there
               are no redundant features.
* n_clusters_per_class: Number of clusters per class. This parameter controls how the data points
                        of each class are grouped. Setting it to 1 means that each class will have
                        its own cluster, resulting in well-separated classes.


 # Visulization

```python
plt.scatter(X[:,0],X[:,1],)
```

```python
x_train , x_test , y_train , y_test  = train_test_split(X,y,shuffle=True,test_size=0.02,random_state=42)
```

# Training the Model

```python
clf = LinearDiscriminantAnalysis()
clf.fit(x_train,y_train)
```

# Predicting the Model

```python
y_pred = clf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
plot_decision_regions(X,y,clf=clf)
```
