# Importing library

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv('Naive-Bayes-Classification-Data.csv')
df
```

# Data Visulization 

```python
plt.scatter(df.iloc[: ,0:1] , df.iloc[: ,1:2])
plt.xlabel('glucose')
plt.ylabel('bloodpressure')
plt.title('Diabetes Data Set')
plt.grid()
```

```python
x_train , x_test , y_train , y_test = train_test_split(df.iloc[:,0:2].values , df.iloc[: ,2:3].values , test_size=0.2 , random_state=42)
```

* Training of Model 

```python
NB_model = GaussianNB()
NB_model.fit(x_train,y_train)
```

* Calculating the Accuracy 

```python
y_pred = NB_model.predict(x_test)
print(f"Accuracy:{r2_score(y_test, y_pred)}")
```

* Predict Known Data 

```python
pred = NB_model.predict(np.array([40,90] ,ndmin=2))
print(f"Prediction of 40 Glucose & 90 Bloodpressure : {pred}")
pred2 = NB_model.predict(np.array([90,40] ,ndmin=2))
print(f"Prediction of 90 Glucose & 40 Bloodpressure : {pred2}")
```
